```python
import pandas as pd

# 创建示例 DataFrame
data = {
   'movie_id': [1, 1, 2, 2, 2],
    'rating': [3, 4, 5, 2, 4]
}
df = pd.DataFrame(data)

# 使用 groupby 并对 rating 列应用 count 聚合函数
movie_count = df.groupby('movie_id')['rating'].count().rename("movie_interaction_count")
print(movie_count)
print(type(movie_count))

# 转换为 DataFrame
movie_count_df = movie_count.reset_index()
print(movie_count_df)
print(type(movie_count_df))
```

```python
#scikit - learn（通常简称为 sklearn）是一个广泛使用的机器学习库，LabelEncoder 就包含在其中。如果你的开发环境中尚未安装 scikit - learn，则需要使用 pip 进行安装 。
#pip install -U scikit - learn有印象，这个装过
from sklearn.preprocessing import LabelEncoder

# 创建 LabelEncoder 实例
lbe = LabelEncoder()

# 示例分类数据
data = ['apple', 'banana', 'apple', 'cherry']

# 拟合和转换数据
encoded_data = lbe.fit_transform(data)
print(encoded_data)

# 查看标签到编码的映射
print(dict(zip(lbe.classes_, lbe.transform(lbe.classes_))))
```
